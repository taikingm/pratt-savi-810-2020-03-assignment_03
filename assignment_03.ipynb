````python
oil = gdf.loc[gdf["commodity"].isin(["Fuel oils","Crude petroleum","Gasoline"])]
oil.head()
````

````python
sns.boxplot([oil.tons_within, oil.value_within]);
````

````python
sns.set()
sns.relplot(x="tons_out", y="value_out", col="commodity",
            hue="commodity", size="value_in",
            data=oil);
````

````python
fd = gdf.loc[gdf["commodity"].isin(["Meat/seafood"])]
fd.head()
````

````python
sns.set()
sns.relplot(x="tons_out", y="value_out", col="commodity",
            hue="state", size="value_in",
            data=fd);
````

````python
fd10 = pd.DataFrame(fd)
fd10 = fd10.nlargest(10,"value_out")
fd10 = gpd.GeoDataFrame(fd10)
````

````python
sns.set()
sns.relplot(x="tons_out", y="value_out", col="commodity",
            hue="state", size="value_in",
            data=fd10);
````

# 4. Map your Data

#### Create GeoDataFrame Map  `.plot()` to see our the spatial distribution of your data

For this, please **add argument** `figsize=(10, 10)` inside our `.plot( )` function. 

* Add a title to your `plt` as well. _Hint `plt.title(...)`. 

````python
fd.info()
````

````python
fd_int = pd.DataFrame(fd)
fd_int["value_per_ton"] = fd_int["value_per_ton"].astype(str)
fd_int = gpd.GeoDataFrame(fd)
````

````python
fd_int.geometry.total_bounds
````

````python
fd_int_bounds = ([-17, 18.917, -50, 71.35])
````

````python
fd_int_bounds
````

````python
fig, ax = plt.subplots(figsize=(10, 10))

#xlim = (-50, -17)
#ylim = (71.35, 18.917)

#ax.set_xlim(xlim)
#ax.set_ylim(ylim)

fd_int.plot(
    column="value_per_ton",
    cmap="magma",
    ax=ax,
    edgecolor="white",
    legend = True,
    linewidth=0.4,
    )
plt.title("US Food Export by State", fontsize=15, fontname="Palatino")
ax.axis("off")
plt.axis("equal");
````
